```python
import pandas as pd

df_all_suiside = pd.read_excel('/content/OECD.xlsx', index_col = 0)
#df_all_suiside = pd.read_excel('/content/OECD.xlsx')

# OECD 평균 자살률
df_oecd_suiside = df_all_suiside.iloc[12]

# 한국 평균 자살률
df_korea_suiside = df_all_suiside.iloc[5]

# 합침
df_suiside = pd.concat([df_oecd_suiside, df_korea_suiside], axis=1)

df_suiside.to_excel(excel_writer='./graph1.xlsx')
```

```python
import pandas as pd

df = pd.read_excel('/content/graph1.xlsx', index_col = 0)

import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='NanumBarunGothic')

fig, ax = plt.subplots()
ax = sns.lineplot(data = df['OECD'], label = 'OECD평균')
ax = sns.lineplot(data = df['KOREA'], label = '대한민국')
ax.set_title('연도별 자살수')
ax.set_ylabel('')
```
